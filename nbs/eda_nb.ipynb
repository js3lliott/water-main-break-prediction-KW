```python
num_break = {}

for pipe in data['ASSETID']:
    if pipe in num_break:
        num_break[pipe] += 1
    else:
        num_break[pipe] = 1
```

```python
data['NUM_BREAKS'] = data['ASSETID'].map(num_break)
```
